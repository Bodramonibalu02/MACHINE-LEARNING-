```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
dir(housing)
```

```python
housing.feature_names
```

```python
import pandas as pd 
df=pd.DataFrame(housing.data,columns=housing.feature_names)
df.head()
```

```python
df['MedHouseVal']=housing.target
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
#### TRAINING THE DATASET 
```

```python
from sklearn.model_selection import train_test_split

X=df.drop('MedHouseVal',axis='columns')
y=df['MedHouseVal']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#### USING LINEAR REGRESSION
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import time 
start=time.time()
model=LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
end=time.time()
y_pred=model.predict(X_test)
r2_linear=r2_score(y_test,y_pred)
mean_linear=mean_squared_error(y_test,y_pred)
print('r2_score:',r2_linear,'mean_square_error:',mean_linear,'Time:',end-start)
```

```python
#### USING GradientBoostingRegressor
```

```python
from sklearn.ensemble import GradientBoostingRegressor
start=time.time()
model= GradientBoostingRegressor()
model.fit(X_train,y_train)
end=time.time()
model.score(X_test,y_test)
y_pred=model.predict(X_test)
r2_gradient=r2_score(y_test,y_pred)
mean_gradient=mean_squared_error(y_test,y_pred)
print('r2_score:',r2_gradient,'mean_square_error:',mean_gradient,'Time:',end-start)
```

```python
from xgboost import XGBRegressor
start=time.time()
model= XGBRegressor()
model.fit(X_train,y_train)
end=time.time()
model.score(X_test,y_test)
y_pred=model.predict(X_test)
r2_gradient=r2_score(y_test,y_pred)
mean_gradient=mean_squared_error(y_test,y_pred)
print('r2_score:',r2_gradient,'mean_square_error:',mean_gradient,'Time:',end-start)
```
